```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
import math

pi = math.pi
def SE_covariance(x1, x2, l):
    K = np.exp(-cdist(x1,x2)**2/(l**2))
    return K
    
# create data
x = np.linspace(-5,5,200)
x = x.reshape(-1,1)
mu = np.zeros(x.shape)

# compute covariance matrix using squared exponential
K = SE_covariance(x, x, 1)

# now we have the mean function and the covariance function
# the GP is fully described
#f = np.random.multivariate_normal(mu.flatten(),K,5)

for i in range(1, 6):
    K = SE_covariance(x, x, i/2)
    f = np.random.multivariate_normal(mu.flatten(),K,5)
    plt.figure()
    plt.title("l = {0}".format(i/2))
    plt.plot(x, f.T)
# plot the data
plt.plot(x,f.T)
plt.show()
```

As we can see in the five plots showing samples of our GP prior with five different values for l, choosing a higher value for l encodes smoothness in our prior.
